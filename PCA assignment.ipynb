```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv("wine.csv")
wine.describe()
wine.head()
```

```python
wine.data = wine.iloc[:,1:]
wine.data.head()
WINE = wine.data.values
WINE
```

```python
wine_normal = scale(WINE)
```

```python
wine_normal
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
#we can get approximately 90% of the info at pc7
```

```python
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:7], columns=['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6', 'pc7']), wine[['Type']]], axis=1)
finalDf
```

### HIERARCHICAL CLUSTERING 

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(finalDf.iloc[:,0:3])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(df_norm, method="average", metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete = AgglomerativeClustering(n_clusters=7, linkage='complete', affinity= "euclidean").fit(df_norm)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
finalDf['clust']=cluster_labels
finalDf
```

```python
finalDf.iloc[:,0:3].groupby(finalDf.clust).mean()
```

```python
# A total of 7 clusters have been identified
```

### KMeans    

```python
def norm_func(i):
    x= (i-i.min())/(i.max() - i.min())
    return (x)
```

```python
df_norm = norm_func(finalDf.iloc[:,0:3])
df_norm
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_)
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=7)
y_kmeans = clf.fit_predict(df_norm)
#we have identified a total of 7 clusters
```
