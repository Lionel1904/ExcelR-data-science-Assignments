```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crime = pd.read_csv("crime_data.csv")
crime
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(df_norm, method="average", metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete = AgglomerativeClustering(n_clusters=7, linkage='complete', affinity= "euclidean").fit(df_norm)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
crime['clust']=cluster_labels
crime
```

```python
crime.iloc[:,1:].groupby(crime.clust).mean()
```

```python
#I have identified 7 clusters and from the above table we can infer that the crime rate is highest in cluster 0 with a total of 
#approximately 378 cases while crime rate is lowest in cluster 2 with a total of approximately 142 cases
```

```python
data = crime[(crime.clust==0)]
data
# These are the US states in which crime rates are comparatively higher
```

```python
data = crime[(crime.clust==2)]
data
#These are the US states in which crime rates are comparatively lower
```

## KMEANS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
crime = pd.read_csv("crime_data.csv")
crime
```

```python
def norm_func(i):
    x= (i-i.min())/(i.max() - i.min())
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_)
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=7)
y_kmeans = clf.fit_predict(df_norm)
#we have identified a total of 7 clusters
```

```python
y_kmeans
clf.labels_
```

```python
md=pd.Series(y_kmeans)
crime['clust']=md
crime
```

```python
crime.iloc[:,1:6].groupby(crime.clust).mean()
```

```python
crime.plot(x="Murder" ,y="Assault" ,c=clf.labels_,kind="scatter", s=30 ,cmap=plt.cm.coolwarm)
plt.title('Clustering using KMeans')
```

```python
#from the above scatterplot we can infer that the crime rate is highest in cluster 0, while crime rate is lowest in cluster 2
```

## DBSCAN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("crime_data.csv")

print(df.head())
df
```

```python
print(df.info())
```

```python
df1 = df.drop(['Unnamed: 0'], axis=1)
```

```python
array=df1.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=1, min_samples=3)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
pd.set_option("display.max_rows", None)
```

```python
cl
```

```python
dfl = pd.concat([df,cl],axis=1)
dfl
```

```python
plt.style.use('classic')
```

```python
dfl.plot(x="Murder",y="Assault", c=dbscan.labels_ ,kind="scatter" ,s=50, cmap=plt.cm.twilight_shifted_r)
plt.title('Clusters using DBScan')
```

```python
#apart from outliers (label = -1) 4 clusters have been identified (labels = 0,1,2,3)
```

```python
dl = dbscan.labels_
```

```python
import sklearn
sklearn.metrics.silhouette_score(X, dl)
```
